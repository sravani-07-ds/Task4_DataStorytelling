# Business Objective

```python
survival_rate = df['Survived'].mean() * 100

print("Overall Survival Rate:", round(survival_rate, 2), "%")
```

```python
gender_survival = df.groupby('Sex')['Survived'].mean() * 100

print(gender_survival)
```

```python
class_survival = df.groupby('Pclass')['Survived'].mean() * 100

print(class_survival)
```

## Data Story

The analysis shows that survival was strongly influenced by gender and passenger class.

Female passengers had a significantly higher survival rate than male passengers.

Passengers travelling in first class had better survival chances compared to second and third class passengers.

These findings suggest that social and economic factors played an important role during the disaster.

# Hypothesis Testing

H0: Male and female passengers had the same survival rate.

H1: Male and female passengers had different survival rates.

# STEP - 9
# CREATE GROUPS

```python
male = df[df['Sex'] == 'male']['Survived']

female = df[df['Sex'] == 'female']['Survived']
```

# STEP - 10
# T-TEST

```python
t_stat, p_value = stats.ttest_ind(
    male,
    female,
    equal_var=False
)

print("T Statistic =", t_stat)
print("P Value =", p_value)
```

# STEP - 11
# DECISION

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
    print("There is a significant difference in survival rates.")
else:
    print("Fail to Reject Null Hypothesis")
    print("No significant difference found.")
```

# Interpretation

The p-value obtained from the t-test is compared against the significance level of 0.05.

If the p-value is less than 0.05, the null hypothesis is rejected.

This indicates that survival rates differed significantly between male and female passengers.

Therefore, gender was an important factor influencing survival on the Titanic.
